# Imagenes para Tesis

```python
import networkx as nx
import community
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('seaborn-poster')
```

# Imagen red aleatorio Erdos-Renyi

```python
G = nx.fast_gnp_random_graph(100,0.1)
nx.draw_random(G)
```

```python
sns.kdeplot(pd.Series([i[1] for i in G.degree()]),  shade=True, cut=3)
plt.title("Distribución del grado en una red Erdös-Renyi")
plt.ylabel("Probabilidad de tener grado k")
plt.xlabel("Grado")
```

# Red Albert-Barabassi

```python
G = nx.barabasi_albert_graph(60,5)
nx.draw_spring(G)
```

```python
sns.kdeplot(pd.Series([i[1] for i in G.degree()]),  shade=True, cut=2)
plt.title("Distribución del grado en una red Barabasi-Albert")
plt.ylabel("Probabilidad de tener grado k")
plt.xlabel("Grado")
```

# Grafo Completo

```python
plt.style.use("default")
```

```python
f = plt.figure(figsize=(18,3))
for i in range(1,8):
    G = nx.complete_graph(i)
    ax = f.add_subplot(int('17'+str(i)))
    nx.draw_circular(G)
    plt.title("$K_%s$" % i)

plt.tight_layout()
```

```python
sns.kdeplot(pd.Series([i[1] for i in G.degree()]),  shade=True, cut=3)
plt.title("Distribución del grado en una red Erdös-Renyi")
plt.ylabel("Probabilidad de tener grado k")
plt.xlabel("Grado")
```

# Transicion de Red Watts-Strogatz

```python
f = plt.figure(figsize=(18,5))
for i,j in enumerate([0,0.5,1]):
#     print(i)
    G = nx.watts_strogatz_graph(12,4,j)
    ax = f.add_subplot(int('13'+str(i+1)))
    nx.draw_circular(G)
    plt.title("n=12, k=4, p=%s" %j, fontsize=16)
    if i == 0:
        plt.text(-0.35,-1.2,"Grafo Regular", fontsize=15)
    elif i == 1:
        plt.text(-0.55,-1.2,"Grafo Mundo Pequeño",fontsize=15)
    else:
        plt.text(-0.45,-1.2,"Grafo Aleatorio", fontsize=15)
```
