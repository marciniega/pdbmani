```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('seaborn-poster')
```

```python
def df_ss_rmsd_frame(path):
    df = pd.read_csv(path)
    df['frame'] = df.proteinaA_proteinaB.str.split('_').str[-1].astype(int)
    df.sort_values('frame',inplace=True)
    df.reset_index(drop=True,inplace=True)
    return(df)
```

```python
ddc = df_ss_rmsd_frame('../result_dimer_vs_dimer_con_cofact.csv')
dds = df_ss_rmsd_frame('../result_dimer_clean_dimer_sin_cofact.csv')
dtc = df_ss_rmsd_frame('../result_tetra_clean_vs_tetra_con_cofact.csv')
dts = df_ss_rmsd_frame('../result_tetra_ref_vs_tetra_sin_cofact.csv')
```

```python
ax = plt.subplot(221)
ddc.plot(x='frame',y='SO',style='.--', 
                 title='SS vs frame en dímero con cofactor',
                 ms=16, legend=False, ax=ax ,figsize=(20,10))
plt.hlines(df.SO.mean(),df.frame.min(),df.frame.max(), color='r')
plt.ylabel('SS')
plt.legend(('SS','SS promedio'))

ax = plt.subplot(222)
dds.plot(x='frame',y='SO',style='.--', 
                 title='SS vs frame en dímero con cofactor',
                 ms=16, legend=False, ax=ax )
plt.hlines(df.SO.mean(),df.frame.min(),df.frame.max(), color='r')
plt.ylabel('SS')
# plt.legend(('SS','SS promedio'))

ax = plt.subplot(223)
dtc.plot(x='frame',y='SO',style='.--', 
                 title='SS vs frame en tetrámero con cofactor',
                 ms=16, legend=False, ax=ax )
plt.hlines(df.SO.mean(),df.frame.min(),df.frame.max(), color='r')
plt.ylabel('SS')
# plt.legend(('SS','SS promedio'))

ax = plt.subplot(224)
dts.plot(x='frame',y='SO',style='.--', 
                 title='SS vs frame en tetrámero sin cofactor',
                 ms=16, legend=False, ax=ax )
plt.hlines(df.SO.mean(),df.frame.min(),df.frame.max(), color='r')
plt.ylabel('SS')
# plt.legend(('SS','SS promedio'))

plt.tight_layout()
plt.savefig('result_ss_vs_frame_dimer_tetra.jpg');
```

```python
ax = plt.subplot(221)
ddc.plot(x='frame',y='RMSD',style='.--', 
                 title='RMSD vs frame en dímero con cofactor',
                 ms=16, legend=False, ax=ax ,figsize=(20,10))
plt.hlines(df.RMSD.mean(),df.frame.min(),df.frame.max(), color='r')
plt.ylabel('RMSD')
plt.legend(('RMSD','RMSD promedio'))

ax = plt.subplot(222)
dds.plot(x='frame',y='RMSD',style='.--', 
                 title='V vs frame en dímero con cofactor',
                 ms=16, legend=False, ax=ax )
plt.hlines(df.RMSD.mean(),df.frame.min(),df.frame.max(), color='r')
plt.ylabel('RMSD')

ax = plt.subplot(223)
dtc.plot(x='frame',y='RMSD',style='.--', 
                 title='RMSD vs frame en tetrámero con cofactor',
                 ms=16, legend=False, ax=ax )
plt.hlines(df.RMSD.mean(),df.frame.min(),df.frame.max(), color='r')
plt.ylabel('RMSD')

ax = plt.subplot(224)
dts.plot(x='frame',y='RMSD',style='.--', 
                 title='RMSD vs frame en tetrámero sin cofactor',
                 ms=16, legend=False, ax=ax )
plt.hlines(df.RMSD.mean(),df.frame.min(),df.frame.max(), color='r')
plt.ylabel('RMSD')

plt.tight_layout()
plt.savefig('result_RMSD_vs_frame_dimer_tetra.jpg');
```
